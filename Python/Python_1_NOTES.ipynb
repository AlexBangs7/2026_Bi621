# Python is a "Dynamically Typed" (or loosely typed) Language

- Type checking happens when the program is run (rather than compiled)
- Variables can change types


````python
a = 15
print(type(a))

a = "15"
print(type(a))

a*a
````

# Contrast with C++ (a "Statically Typed") Language
 - Type checking is done at compile time
 - Variables cannot change types

# Python type annotations
- Introduced in Python 3.0
- TOTALLY OPTIONAL (as far as Python is concerned)

````python
a: int = 15
print(type(a))

b: str = "15"
print(type(b))
````

````python
#Note! Python totally ignores the type annotations!!!
a: str = 15
print(type(a))

b: int = "15"
print(type(b))
````
